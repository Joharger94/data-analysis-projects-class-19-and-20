```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to 
## get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.

release_years = np.sort(df["year"].unique().astype(int))
book_sum_by_year = np.sort(df.groupby("year")["bookID"].count())

plt.plot(release_years, book_sum_by_year)
plt.title("Amount Of Books Published By Year")
plt.xlabel("Year")
plt.ylabel("Number Of Books Published")
```

```python
# Use your Python skills to create a visualization here.
years = np.sort(df["year"].unique().astype(int))
avg_book_page_count_by_year = np.sort(df.groupby("year")["  num_pages"].mean())
book_ratings_data = pd.DataFrame({'Years': years, 'Book Page Counts': avg_book_page_count_by_year})
sns.lineplot(data=book_ratings_data, x="Years", y="Book Page Counts")
```
